Імпортуємо потрібні бібліотеки: pandas i numpy

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('EV_cars.csv')
df
```

Порахувати середнє значення батареї, використовуючи масив numpy

```python
mid_b = round(np.array(df['Battery']).mean(), 1)
mid_b
```

Додати стовпець, де для кожного автомобіля порахувати, скільки разів він проїде від гуртожитку до ЛНУ ім. Івана Франка

```python
distance = 5.4 # кілометри
df['to_LNU_count'] = round(df['Range'] / distance, 1)
df.head()
```

Вивести топ 5 автомобілів марки Tesla за ціною

```python
search = 'Tesla'
tesla_df = df[df['Car_name'].str.startswith(search)]
indexes = [i for i in tesla_df['Price.DE.'].nlargest(n=5).index] # [107, 17, 103, 22, 14]

df.iloc[indexes]
```

Обчисліть кореляцію між запасом ходу та ціною автомобілів

```python
df = df.dropna()
range_data = df['Range']
price_data = df['Price.DE.']

correlation = np.corrcoef(range_data, price_data)
print(f'{round(correlation[1][0], 2)}, Це означає, що кореляція між запасом ходу і ціною автомобіля велика')
```

Знайдіть електромобілі з найвищим співвідношенням запасу ходу до ціни

```python
battery_type = df['Battery']

average_price_by_battery_type = df.groupby('Battery')['Price.DE.'].mean()

a = []
for battery_type, average_price in average_price_by_battery_type.items():
    a.append(f'Ємність батареї: {battery_type}, середня ціна: {average_price:.2f} євро')

for x in a[-5:]:
    print(x)
```
